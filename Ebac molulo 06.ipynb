```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o estilo do seaborn para melhorar a apresentação dos gráficos
sns.set(style="whitegrid")

# Criando subplots para os gráficos de barras
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Gráfico de barras para APGAR1
sns.countplot(x='APGAR1', data=sinasc_data, ax=axs[0, 0])
axs[0, 0].set_title('APGAR1')

# Gráfico de barras para APGAR5
sns.countplot(x='APGAR5', data=sinasc_data, ax=axs[0, 1])
axs[0, 1].set_title('APGAR5')

# Gráfico de barras para SEXO
sns.countplot(x='SEXO', data=sinasc_data, ax=axs[1, 0])
axs[1, 0].set_title('SEXO')

# Gráfico de barras para GESTACAO
sns.countplot(x='GESTACAO', data=sinasc_data, ax=axs[1, 1])
axs[1, 1].set_title('GESTACAO')

# Gráfico de barras para ESTCIVMAE
sns.countplot(x='ESTCIVMAE', data=sinasc_data, ax=axs[2, 0])
axs[2, 0].set_title('ESTCIVMAE')

# Removendo o subplot vazio
fig.delaxes(axs[2, 1])

# Ajustando o layout
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculando a média do APGAR5 por CONSULTAS
avg_apgar5_by_consultas = sinasc_data.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Configurando o estilo do seaborn
sns.set(style="whitegrid")

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=avg_apgar5_by_consultas, palette="viridis")

# Adicionando rótulos e título
plt.xlabel('Número de Consultas Pré-Natais')
plt.ylabel('Média do APGAR5')
plt.title('Média do APGAR5 por Número de Consultas Pré-Natais')

# Exibindo o gráfico
plt.show()
```
